```python
import json  
import pandas as pd  
from pandas.io.json import json_normalize  
import numpy as np

with open('result.json') as f: 
    d = json.load(f) 
  
#lets put the data into a pandas df 
data = json_normalize(d) 

#get columns
for col in data.columns:   print(col)   
```
