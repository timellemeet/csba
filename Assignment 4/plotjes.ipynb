```python
import json  
import pandas as pd  
from pandas.io.json import json_normalize  
import numpy as np

with open('result.json') as f: 
    d = json.load(f) 
  
#lets put the data into a pandas df 
data = json_normalize(d) 

#get columns
for col in data.columns:   print(col)   
```

```python
topn = 10

traits = [
    "data.receptiviti_scores.percentiles.assertive",
    "data.receptiviti_scores.percentiles.imaginative"
]


tables = []
keys = []

for t in traits:
    label =  t.split('.')[-1].capitalize()
    tdf =  data.sort_values(by=[t], ascending=False)[["name",t]].reset_index().head(topn).round(2)
#     tdf["index"] = tdf.index+1
    tdf = tdf.rename(columns={"name": "Person", t: "Score"})
    
    tables.append(tdf[["Person", "Score"]])
    keys.append(label)
    
results = pd.concat(tables, keys=keys, axis=1)
results.insert(0,"No.", results.index+1) 
display(results)
print(results.to_latex(index=False))


```
